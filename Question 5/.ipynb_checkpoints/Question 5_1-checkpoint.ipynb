```python
import numpy as np
import itertools
from random import shuffle
import time

start = time.time()

###########################################################################################

# ============= Generating an ordered list =============
## Using the ordered list from Question 1
## Each element in generated list will be reshaped to a L**2 matrix later on for each run

def practice_5_ordered(L):
    
    def ordered_placement(size, count):
        for positions in itertools.combinations(range(size), count):
            p = ['B'] * size
            for i in positions:
                p[i] = 'R'
            yield p
    total_beads = L**2
    red_beads = int(total_beads/2)
    beads_placement = list(ordered_placement(total_beads, red_beads))
#     print(len(beads_placement))

    return beads_placement


# ============= Neighbour function =============
## Returns neighbouring values
def neighbour(row, col, img = img_padded):
    top = img[row - 1, col]
    left = img[row, col - 1]
    right = img[row, col + 1]
    bottom = img[row + 1, col]

## Fitting all the neighbouring values into an array for checking later
    neighbour_array = [top, left, right, bottom]
    return neighbour_array

###########################################################################################

L = 5

# print(len(practice_5_ordered(L)))
# print(practice_5_ordered(L))

ordered_list = practice_5_ordered(L)
# print(ordered_list)
# print(np.size(ordered_list))

# Initiate penalty_list, same dimensions as ordered_list
penalty_list = np.zeros(np.size(ordered_list, 0), dtype = int)
# print(penalty_list)

# Initiate img_output
img_output = np.zeros([L, L], dtype = int)
# print(img_output)

###########################################################################################

for x in range(len(ordered_list)):
# for x in range():

# ============= Reshape list to matrix for 4-connectivity test =============
    np_img = np.reshape(ordered_list[x], (L, L))
#     print(np_img)

# ============= Adding 0s around the image =============
## To facilitate the calculations later

## Adding zeroes to sides
    img_addcol = np.c_[np.zeros(L, dtype = int), np_img, np.zeros(L, dtype = int)]
#     print (img_addcol)

## Adding zeroes to top and bottom
    img_padded = np.r_[[np.zeros((L + 2), dtype = int)], img_addcol, [np.zeros((L + 2), dtype = int)]]
#     print(img_padded)

# ============= 4-connectivity =============
## Going through the rows first before going through the columns

# Initiating penalty/Reset penalty for each run
    penalty = 0

    for col in range(1, L + 1):
        for row in range(1, L + 1):

            # If value in cell is B, check values of all 4 neighbour cells
            if img_padded[row, col] == 'B':
#                 print(img_padded[row, col])
                # Need to feed img_padded into neighbour function, else every run will give the same values
                neighbour_value = neighbour(row, col,img_padded)
#                 print(neighbour_value)
                
                # Add 1 to penalty if neighbour cell is also B
                for i in range(len(neighbour_value)):
                    if neighbour_value[i] == 'B':
                        penalty += 1
#                         print(penalty)
            
            # Else value in cell is R, check values of all 4 neighbour cells
            else:
#                 print(img_padded[row, col])
                # Need to feed img_padded into neighbour function, else every run will give the same values
                neighbour_value = neighbour(row, col,img_padded)
#                 print(neighbour_value)
                
                # Add 1 to penalty if neighbour cell is also B
                for i in range(len(neighbour_value)):
                    if neighbour_value[i] == 'R':
                        penalty += 1
#                         print(penalty)
    
    # Updates penalty_list after each run
    # penalties divided by 2 cause it was counted twice 
    penalty_list[x] = penalty/2
#     print(penalty_list)

# Find infices of min element
min_penalty = np.argmin(penalty_list)
print(min_penalty)
print(penalty_list[min_penalty])
print(np.reshape(ordered_list[min_penalty], (L, L)))

img_output = np.reshape(ordered_list[min_penalty], (L, L))
# print(img_output)

img_export = np.savetxt("output_question_5_1.txt", img_output, fmt = '%s')

end = time.time()
print('Time taken for fun program: ', end - start)
```
