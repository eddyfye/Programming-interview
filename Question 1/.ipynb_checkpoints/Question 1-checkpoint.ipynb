```python
import numpy as np
import itertools
from random import shuffle
```

```python
def practice_1_rand(m, n, summed_number):
    
###     # Creating a matrix with m and n values

###    Initial Matrix setup 1
#     a_column = np.array([i+1 for i in range(m)])
# #     print(a_column)
#     a_row = a_column.reshape(m,1)
# #     print(a_row)
#     a_ones = np.ones((1, n), dtype = int)
# #     print(a_ones)
#     matrix_a = a_row * a_ones
# #     print(matrix_a)


###    # Matrix setup 2 - (faster way)
###    # Setup a matrix of ones and change the row values using for loop 
    matrix_a = np.ones([m,n], dtype = int)
#     print(matrix_a)
    for i in range(m):
        matrix_a[i,:] = i + 1
        i += 1
#     print(matrix_a)


###    # Generate a list of operations for calculation
###    # Randomize the list after every run and check answer with summed_number
###    # A failed list is used to prevent repeats of the same order of operations

    # Creates a list of string of R for number of m, followed by D for number of n
    # len - 2 as operations always 2 less than numbers
    combined_list = ['R'] * (n - 1) + ['D'] * (m - 1)
#     print(combined_list)

    random_list = combined_list.copy()
#     print(random_list)

    # Initiate with first number and failed_list
    combined_answer = matrix_a[0][0]
    failed_list = []

    # Loop till combined_answer = summed_number 
    while combined_answer != summed_number:
        # Reset combined_answer, column and row for every loop
        column_num = 0
        row_num = 0
        combined_answer = matrix_a[0][0] 
    
        # First loop gets passed through directly without randomization
        if failed_list != []:
            while random_list in failed_list:
                # shuffle random_list for random R and D
                shuffle(random_list)
#                 print('new random = ' + str(random_list))
                if random_list in failed_list:
                    break
        
        # Goes through the random_list and sum values according to R or D
        for i in range(len(random_list)):
            if random_list[i] == 'R':
                column_num += 1
#                 print(column_num)
            else:
                row_num += 1
#                 print(row_num)
            
#             print(row_num, column_num)
#             print('+ ' + str(a[row_num][column_num]))
            combined_answer += matrix_a[row_num][column_num]
#             print(combined_answer)
        
        # If random_list does not give the desired answer, add to failed_list
        # Using failed_list to prevent repeat for future loops
        if combined_answer != summed_number:
            failed_list.append(random_list.copy())
#             print('Failed list = ' + str(failed_list))
    
    # Using join to convert to a string of operations for print
    operation = ''.join(random_list)
#     print(operation)
    print(str(combined_answer) + ' ' + operation)
    
    return [str(combined_answer) + ' ' + operation]
```

```python
###    # Generate a full list of all potential operations for calculation
###    # as compared to the randomizing after every run

def practice_1_ordered(m, n, summed_number):
    
###    # Matrix setup 2 - (faster way)
###    # Setup a matrix of ones and change the row values using for loop 
    matrix_a = np.ones([m,n], dtype = int)
#     print(matrix_a)
    for i in range(m):
        matrix_a[i,:] = i + 1
        i += 1
#     print(matrix_a)


    # Creates a list of string of R for number of m, followed by D for number of n
    combined_list = ['R'] * (n - 1) + ['D'] * (m - 1)
#     print(combined_list)

##    # Goes through each tuple to see if it has been repeated, if not, add to list.
##    # Slowest method.
##    # Generate a random list of operations as tuples of tuples
#     random_tuple = itertools.permutations(combined_list)
#     random_no_dup = [] 
#     [random_no_dup.append(x) for x in random_tuple if x not in random_no_dup] 
#     print(random_no_dup)


##    # Using set to remove duplicates, works well for small list
##    # Super slow for longer ones.
##    # Generate a random list of operations as tuples of tuples
#     random_tuple = itertools.permutations(combined_list)    
#     random_set = set(random_tuple)    
#    # Converts the set of unique operations to a list of lists
#    random_no_dup = [list(i) for i in random_set]
#    print(random_no_dup)


##    # Using ordered_placement to generate a list of no repeats through combinations
##    # Fastest way, but still too slow for large m and n
#     when m and n = 16, size = 155,117,500 and time ~ 4 mins 36s
#     when m and n = 17, size = 601,080,390 and time ~ 1 hr 20 mins
    def ordered_placement(size, count):
        for positions in combinations(range(size), count):
            p = ['D'] * size
            for i in positions:
                p[i] = 'R'
            yield p
    num_RD = m + n - 2
    num_RD_half = int(num_RD / 2)
    random_no_dup = ordered_placement(num_RD, num_RD_half)
    print(len(list(ordered_placement(num_RD, num_RD_half))))
    
    
    
    
    # Initiate with first number and failed_list
    combined_answer = matrix_a[0][0]
    
    # Goes through the full_random_list and check if combined_answer == summed_number
    for i in range(len(random_no_dup)):
        # Reset combined_answer, column and row for every loop
        column_num = 0
        row_num = 0
        combined_answer = matrix_a[0][0]         
    
        # Goes through the random_list and sum values according to R or D, len - 2 as operations always 1 less than numbers
        for x in range(len(random_no_dup[i])):
            if random_no_dup[i][x] == 'R':
                column_num += 1
#                 print(column_num)
            else:
                row_num += 1
#                 print(row_num)
            
#             print(row_num, column_num)
#             print('+ ' + str(a[row_num][column_num]))
            combined_answer += matrix_a[row_num][column_num]
#             print(combined_answer)
        
        print(combined_answer)

        if combined_answer == summed_number:
            break
        

    
    # Using join to convert to a string of operations for print
    operation = ''.join(random_no_dup[i])
#     print(operation)
    print(str(combined_answer) + ' ' + operation)
    
```
