```python
## 3D - Works for example

import numpy as np

l1 = 4
l2 = 3
l3 = 2

grid_test_1 = np.arange(l2 * l1 * l3).reshape(l3, l2, l1)

# print(grid_test_1)

index_value = np.array([[16],
                        [6],
                        [8],
                        [23]])

coord_value = np.array([[2, 0, 1],
                        [3, 1, 0],
                        [1, 1, 0],
                        [2, 2, 0]])


################### Converting index to coordinates ###################

# x1 is column, x2 is row, x3 is 3D

# Creates array for values input later
coordinates_output = np.zeros((np.size(index_value, 0), 3), dtype = int)
# print(coordinates_output)

for i in range(len(index_value)):
    
    # x3 is the remainder after dividing with the number of columns
    x3 = index_value[i] // (l1 * l2)
    remainder = index_value[i] % (l1 * l2)

    # x1 is the remainder after dividing with the number of columns
    if remainder != 0:
        x1 = remainder % l1
#         print(x1)
    else:
        x1 = 0
#         print(x1)
    
    # x1 is the remainder after dividing with the number of columns
    if remainder != 0:
        x2 = remainder // l1
#         print(x2)
    else:
        x2 = 0
#         print(x2)

    coordinates_output[i] = [x1, x2, x3]
#     print(coordinates_output)
    
print(coordinates_output)

################### Converting coordinates to index ###################

# x1 is column, x2 is row, x3 is 3d

# Creates array for values input later
index_output = np.zeros((np.size(coord_value, 0), 1), dtype = int)
# print(coordinates_output)

for i in range(len(coord_value)):
    result_index = coord_value[i][0] + (coord_value[i][1] * l1) + (coord_value[i][2] * (l1 * l2))
#     print(result_index)
    index_output[i] = result_index

print(index_output)
```
